```python
import os
import ccxt
from dotenv import load_dotenv
```

```python
load_dotenv()

# 代理要求：币安允许以下地区的 IP 访问：香港、日本、新加坡、韩国、欧洲多数国家等。测试显示日本可以。

proxies = {
    'http': os.environ.get('PROXY_HOST') + ':' + os.environ.get('PROXY_PORT'),   # 通常是7890或1080/10809
    'https': os.environ.get('PROXY_HOST') + ':' + os.environ.get('PROXY_PORT')   # https也使用http代理即可
}

# 初始化交易所时传入代理配置
exchange = ccxt.binance({
    'apiKey': os.environ.get('BINANCE_API_KEY'),
    'secret': os.environ.get('BINANCE_SECRET'),
    'enableRateLimit': True,
    'proxies': proxies,  # 关键：加入代理设置
    'options': {
        'adjustForTimeDifference': True,
    }
})
```

```python
exchange
```

```python
exchange.load_markets()
# len(exchange.markets)
exchange.markets['BTC/USDT']
```

```python
ticker = exchange.fetch_ticker('BTC/USDT')
ticker
```

```python
order_book = exchange.fetch_order_book('ETH/USDT')
order_book
```

```python
ohlcv = exchange.fetch_ohlcv('BTC/USDT', '1h', limit=10)
for candle in ohlcv:
    print(f"时间: {candle[0]}, 开: {candle[1]}, 高: {candle[2]}, 低: {candle[3]}, 收: {candle[4]}, 量: {candle[5]}")
```

```python
balance = exchange.fetch_balance()
balance
```
